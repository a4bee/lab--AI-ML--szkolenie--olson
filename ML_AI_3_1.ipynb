## MAE

Średni błąd bezwzględny/absolutny (ang. *Mean absolute error*) 

<div style="text-align: centered;">
    <img src="Zdjęcia/Szkolenie_3/MAE.png" alt="drawing" style="width:900:"/>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5, 6])
y = np.array([10, 11, 10, 11, 10, 11])

y_pred = np.mean(y) * np.ones_like(y)

# Obliczanie błędów
mae = np.mean(np.abs(y - y_pred))

# Drukowanie wyników
print(f"MAE: {mae}")

# Wizualizacja punktów danych, linii regresji oraz błędów
plt.figure(figsize=(8, 4))
plt.scatter(X, y, color='blue', label='Punkty danych')
plt.plot(X, y_pred, color='red', linewidth=2, linestyle='--', label='Linia regresji')

# Wizualizacja błędów
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_pred[i]], color='green', linestyle='dotted')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresja z wizualizacją błędów')
plt.legend()
plt.show()
```

MAE = (|11 - 10.5| + |10 - 10.5| + |11 - 10.5| + |10 - 10.5| + |11 - 10.5| + |10 - 10.5|) / 6

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression

# Generowanie losowych punktów danych
np.random.seed(42)
X = 1 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Trenowanie modelu regresji liniowej
model = LinearRegression()
model.fit(X, y)

# Predykcja na podstawie modelu
y_pred = model.predict(X)

# Obliczanie błędów
mae = mean_absolute_error(y, y_pred)

# Drukowanie wyników
print(f"MAE: {mae}")

# Wizualizacja punktów danych
plt.figure(figsize=(8, 4))
plt.scatter(X, y, color='blue', label='Punkty danych')
plt.plot(X, y_pred, color='red', linewidth=2, label='Linia regresji')

# Wizualizacja błędów
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_pred[i]], color='green', linestyle='dotted')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresja liniowa z wizualizacją błędów')
plt.legend()
plt.show()
```

## MSE i RMSE

Średni błąd kwadratowy i pierwiastek średniego błędu kwadratowego (ang. *Mean squared error*, ang. *Root mean squared error*) 

<div style="text-align: centered;">
    <img src="Zdjęcia/Szkolenie_3/mse.png" alt="drawing" style="width:900:"/>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# Generowanie danych
X = np.array([1, 2, 3, 4, 5, 6])
y = np.array([12, 9, 12, 9, 12, 9])

# Przewidywana wartość to średnia y
y_pred = np.mean(y) * np.ones_like(y)

# Obliczanie błędów
errors = y - y_pred
mse = np.mean(errors**2)
rmse = np.sqrt(mse)

# Drukowanie wyników
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")

# Wizualizacja punktów danych
plt.figure(figsize=(8, 4))
plt.scatter(X, y, color='blue', label='Punkty danych')
plt.plot(X, y_pred, color='red', linewidth=2, linestyle='--', label='Linia regresji')

# Wizualizacja błędów
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_pred[i]], color='green', linestyle='dotted')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresja z wizualizacją błędów dla danych prawie pionowych')
plt.legend()
plt.show()
```

MSE = ([12 - 10.5]^2 + [9 - 10.5]^2 + [9 - 10.5]^2 +[9 - 10.5]^2 + [9 - 10.5]^2 + [9 - 10.5]^2) / 6

RMSE = root(MSE)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression

# Generowanie losowych punktów danych
np.random.seed(42)
X = 1 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Trenowanie modelu regresji liniowej
model = LinearRegression()
model.fit(X, y)

# Predykcja na podstawie modelu
y_pred = model.predict(X)

# Obliczanie błędów
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

# Drukowanie wyników
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")

# Wizualizacja punktów danych
plt.figure(figsize=(8, 4))
plt.scatter(X, y, color='blue', label='Punkty danych')
plt.plot(X, y_pred, color='red', linewidth=2, label='Linia regresji')

# Wizualizacja błędów
for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_pred[i]], color='green', linestyle='dotted')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresja liniowa z wizualizacją błędów')
plt.legend()
plt.show()
```

## Gradient prosty

Metoda gradientu prostego (ang. *gradient descent*) stanowi bardzo prosty algorytm optymalizacyjny służący do znajdowania optymalnych rozwiązań dla bardzo szerokiej gammy problemów. Ogólna koncepcja problemu polega na iteracyjnym poprawianiu parametrów w celu minimalizowaniu funkcji kosztu. Należy wyobrazić sobie kulkę w misce, która turla się na sam dół. Algorytm mierzy lokalny gradient funkcji w odniesieniu do wektora parametrów (theta), a następnie porusza się w kierunku malejącego gradientu. Dla wartości gradientu 0 mamy minimum/maksimum funkcji.

## Gradient wsadowy prosty

Aby zaimplementować metodę gradientu prostego (ang. *batch gradient descent*) należy obliczyć gradient funkcji kosztu wobec każdej cechy modelu (należy policzyć jak bardzo funkcja kosztu ulegnie zmianie, jeśli tylko trochę zmodyfikuje się daną cechę). Proces tego wyliczania nazywa się **pochodą cząstkową** (ang. *partial derivative*). Można sobie to wybrazić jako: jesteście w górach podczas nocy i jest tak ciemno, że nic nie widać. Jeśli chcecie zejść z góry to intuicyjne wydaje się kroczenie nogami w tę stronę gdzie będzie spadek/obniżenie terenu. Możemy iść na północ, południe, wschód i zachód. Licząc pochodną cząstkową sprawdzamy dla każdego kierunku, jak wielkie jest obniżenie terenu na północ, południe, wschód i zachód. Wybieramy ten kierunek, gdzie spadek jest największy.

<br>

Nazwa algorytmu wzięła się od tego, że w każdej iteracji obliczenia sią tworzone dla całego zbioru X (zbioru uczącego/treningowego)

<br>

Warto pamiętać, że z matematycznego punktu widzenia gradient pokaże na miejsce wchodzenia w górę, dlatego należy skierować się w przeciwną stronę (od wartości parametrów należy odjąć wektor gradientów)

theta_next = theta - learning_rate * wektor_gradientów

## Stochastyczny spadek wzdłuż gradientu

Główny problem wsadowego gradientu prostego jest wykorzystanie pełnego zbioru danych uczących w każdej iteracji, przez co wraz ze wzrostem rozmiaru uczącego spada jego szybkość. Jego przeciwieństwo stanowi algorytm **stochastycznego spadku wzdłuż gradientu** (ang. *stochastic gradient descent*) w którym podczas każdej iteracji dobierana jest losowa próbka ucząca, za której pomocą wyliczane są gradienty. Dzięki temu algorytm jest szybszy, ale w wyniku stochastycznej (losowej) natury algorytm ten będzie bardziej chaotyczny. Funkcja kosztu nie schodzi tu tak łagodnie, lecz przeskakuje losowo dążąc do minimum (w którym również będzie skakać). Uzyskanie rozwiązanie będzie co najwyżej dobre, a nie optymalne.

<br>

W celu minimalizacji zjawiska skoków należy stopniowo zmniejszać wartość współczynika uczenia (na początku preferowane będzie poszukiwanie minimum, a wraz ze zwiększeniem iteracji model stopniowo zacznie szukać dokładnego minimum). Podejście te w optymalizacji nazywane jest symulowanym wyżarzaniem (ang. *simulated annealing*). Można sobie wyobrazić rozgrzadny do czerwoności metal, który jest plastyczny, a w miarę upływającego czasu (chłodzenia) staje się twardy. Funkcja określająca współczynnik uczenia w każdym przebiegu nazywana jest harmonogramem uczenia (ang. *learning schedule*).

## Obciążenie a wariancja 

**Obciążenie** (ang. *bias*) - ta składowa błędu uogólnienia wynika z nieprawidłowych założeń, np. że dane są liniowe, podczas gdy w rzeczywistości są opisane funkcją kwadratową. Model o wysokiej wartości obciążenia ma największą szansę na niedotrenowanie wobec danych uczących.

<br>

**Wariancja** (ang. variance) - ta składowa wynika z nadmiernej czułości modelu na drobne wahania w wartościach danych uczących. Model cechujący się dużą liczbą stopni swobody (np. wielomianowy) prawdopodobnie będzie miał dużą wariancję, a przez to będzie przetrenowany wobec danych uczących.

<br>

**Błąd nieredukowalny** (ang. *irreducible error*) ta składowa stanowi konsekwencję zaszumienia danych. Jedynym sposobem pozbycia się jej jest oczyszczenie danych.

<br>

Zwiększanie złożoności modelu zazwyczaj zwiększa jego wariancję i zmniejsza obciążenie. Odwrotny proces następuje w przypadku uproszczenia modelu, dlatego należy tu mówić o kompromisie.

Dobrym sposobem zmiejszania przetrenowania modelu jest jego regularyzacja (tzn. ograniczenie). W przypadku modelu liniowego regularyzacja jest przeważnie osiągana poprzez ograniczanie wag modelu (wpływu cechy na predykcję).

## Regresja grzbietowa

Regresja grzbietowa/siodłowa (ang. *ridge regression*) stanowi modyfikację regresji liniowej. Do funkcji kosztu dodaje się człon regularyzacyjny w postaci:

$\alpha \cdot \sum_{i=1}^{n} \theta_i^2$

W ten sposób model jest nie tylko do dopasowania się do danych ale również do utrzymania jak najmniejszych wag

Funkcja kosztu w metodzie grzbietowej ma następującą postać:

$J(\theta) = MSE(\theta) + \alpha * 1/2 * \sum_{i=1}^{n} \theta_i^2$

Ważne jest, aby przed przerpowadzeniem regresji grzbietowej przeskalować dane za pomocą klasy StandardScaler. Standaryacja to transofmacja danych polegająca na: odcjęciu od każdej wartości wartość średnią (czyli średnia z próbek zawsze wynosi 0), a następnie należy je podzielić przez wariancję (dzięki czemu wynikowy rozkład ma wariancję jednostkową). Standaryzacja jest mniej podatna na elementy odstające. 

Innym typem skalowania jest normalizacja (klasa MinMaxScaler), która sprawia, że wartości mieszczą się w zakresie 0 - 1, np. dla wartości z przedziału 0 - 15 powstałby przedział 0 - 0.15.

## Regresja LASSO

Regresja LASSO (ang. *least absolute shrinkage and selection operator regression*), w wolnym tłumaczeniu operator najmniejszej bezwzględnej redukcji i wyboru. Do funkcji kosztu dodaje się człon regularyzacyjny w postaci:

$\alpha \cdot \sum_{i=1}^{n} |\theta_i|$

Ważną własnością regresji LASSO jest to, że dąży ona do całkowitej eliminacji wag w mniej istotnych cechach

Funkcja kosztu w metodzie LASSO ma następującą postać:

$J(\theta) = MSE(\theta) + \alpha \cdot \sum_{i=1}^{n} |\theta_i|$

## Regresja Elastic Net

Metoda elastycznej siatki (ang *elastic net*) stanowi pośrednią metodę między regresją grzbietową, a regresją LASSO. Człon reglaryzacyjny tworza prosta kombinacja obu członów. Możemy nim sterować za pomocą współczynnika proporcji r.

Funkcja kosztu w metodzie elastic net ma następującą postać:

$J(\theta) = MSE(\theta) + r * \alpha \cdot \sum_{i=1}^{n} |\theta_i| + \alpha * \frac{1 - r}{2} * \sum_{i=1}^{n} \theta_i^2$

W sklearn używając klasy ElsaticNet ustawiamy paramet alfa i l1_ratio, wartość parametru l2 to 1-l1

## Softmax

Można uogólnić model regresji logistycznej do bezpośredniej obsługi wielu klas bez konieczności uczenia i łączenia wielu klasyfikatorów binarnych. Tego typu model nosi nazwę regresji softmax lub wielomianowej/wielorakiej regresji logistycznej (ang. *multinomial logistic regresion*). W skrócie model będzie obliczał dla danej wartości prawdopodobieńswo przynależności do danej klasy.
