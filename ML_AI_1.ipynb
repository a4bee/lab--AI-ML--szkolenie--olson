```python
import matplotlib.pyplot as plt
import numpy as np

# Seed the random number generator for reproducibility
np.random.seed(42)

# Generate random x values
X = 10 * np.random.rand(100)

# Generate random y values
Y = 2.5 * X + np.random.randn(100) * 5

# Create 3 linear functions
y1 = 0.5 * X + 1
y2 = -0.4 * X + 10
y3 = 2 * X + 12.5

# Create a figure with a specific size
plt.figure(figsize=(10, 6))

# Scatter plot of the data
plt.scatter(X, Y, color='blue')

# Plot the three linear regression lines
plt.plot(X, y1, color='yellow', label='y1')
plt.plot(X, y2, color='red', label='y2')
plt.plot(X, y3, color='green', label='y3')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# Seed the random number generator for reproducibility
np.random.seed(42)

# Generate random x values
X = 10 * np.random.rand(100)

# Generate random y values
Y = 2.5 * X + np.random.randn(100) * 5

# Reshape from rows to columns

X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)

# Create Linear Regression model
model = LinearRegression()

# Fit model into data
print("Fitting the model...")
model.fit(X,Y)

# Predict y values with model
Y_pred = model.predict(X)

# Calculate Mae
mae = mean_absolute_error(Y, Y_pred)
print(f"MAE: {mae}")

# Calculate STD
std = np.std(Y - Y_pred)
print(f"STD: {std}")

plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue')
plt.plot(X, Y_pred, color='red', label='Linear regresion')
plt.xlabel("Zadowolenie")
plt.ylabel("PKB per capita")
plt.legend()

plt.show()

```

```python
X_new = [[8.8]]
print(model.predict(X_new))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error

# Seed the random number generator for reproducibility
np.random.seed(42)

# Generate random x values
X = 10 * np.random.rand(100)

# Generate random y values
Y = 2.5 * X + np.random.randn(100) * 5

# Reshape from rows to columns
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)

# Create K-Nearest Neighbors Regressor model
model = KNeighborsRegressor(n_neighbors=3)

# Fit model to data
print("Fitting the model...")
model.fit(X, Y)

# Predict y values with model
Y_pred = model.predict(X)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(Y, Y_pred)
print(f"MAE: {mae}")

# Calculate Standard Deviation (STD)
std = np.std(Y - Y_pred)
print(f"STD: {std}")

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Actual data')
plt.scatter(X, Y_pred, color='red', label='Predicted values', marker='x')
plt.xlabel("Feature X")
plt.ylabel("Target Y")
plt.legend()
plt.title("KNN Regression Results")
plt.show()
```

```python
X_new = [[8.8]]
print(model.predict(X_new))
```

Zjawisko przetrenowania występuje, gdy model jest zbyt skomplikowany w prównaniu do ilości danch uczących. W takich przypadkach możliwe są następujące rozwiązania:
    - uproszczenie modelu
    - uzyskanie większej ilości danych uczących
    - zmniejszenie zaszumienia danych
    - zmniejszanie czasu nauki

Ograniczenie modelu w celu jego uproszczenia i zmniejszenia ryzyka przetrenowania nosi nazwę regularyzacji. Stopień regularyzacji można kontrolować za pomocą hiperparametrów (ang. hyperparameters). Najprościej mówiąc są to parametry algorytmu uczącego (a nie całego modelu)

Jakie parametry można ograniczyć w regresji liniowej?

<center>
<font size="6"> Niedotrenowanie danych uczących (ang underfitting) (ang. overfitting) </font>
</center>

Przeciwieństwo przetrenowania. Model jest zbyt prosty, aby wyuczyć się struktur danych uczących (rzeczywistość jest dużo bradziej skomplikowana). Przykład poniżej.
Sposoby na niedotrenowanie:
    - wybór bardziej złożonego modelu
    - dołączenie większej ilośći cech (inżynieria cech)
    - zmniejszenie ograniczeń midelu

<br><br>

Jak wykryć przetrenowanie?

<div style="text-align: center;">
  <img src="Zdjęcia/Szkoleie_1/over_fit_plot.png" alt="drawing" style="width:900px;"/>
</div>

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Seed the random number generator for reproducibility
np.random.seed(42)

X = 10 * (np.random.rand(100) - 0.5)
Y = (2.5 * np.power(X,2)) + np.random.rand(100)

X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)

model = LinearRegression()
model.fit(X,Y)

Y_pred = model.predict(X)

plt.figure(figsize=(10,6))
plt.scatter(X,Y)
plt.plot(X,Y_pred,color='red')
```

# Testowanie i ocenianie 

Zbiory danych zwykle dzielone są zbiór uczący (ang. training set) oraz zbiór testowy (ang. test set). Model trenuje się na zbiorze treningowy, a sprawdza przy użyciu zbioru testowego. Dzięki zbiorowi testowemu możemy określić tzw. współczynnik generalizacji, czyli określić jak dobrze model spisuje się wobec nieznanych danych. Jeśli jego wartość dokładność dla zbioru uczącego jest wysoka, ale za to duża dla zbioru testowego to oznacza to, że model jest przetrenowany. Wartości hiperparametrów należy dobierać do **zbioru testowego**. Aby najlepiej określić dokładność modelu używa się metody zwanej sprawdzianem krzyżowym lub kroswalidacją (ang. cross-validation). Polega on na podziale całego zbioru danych na k składek, gdzie przez k iteracji będzie zmieniać się zbiór uczący oraz treningowy (każda składka raz zostanie zbiorem uczącym). Zazwyczaj zbiór danych dzieli się w stosunku 80:20 (80% dla zbioru uczącego i 20% dla zbioru testowego)
